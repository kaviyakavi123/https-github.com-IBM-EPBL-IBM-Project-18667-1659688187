## 9. Generate the series of dates from 1st Jan, 2023 to 10th Feb, 2023

```python
import pandas as pd

# creating range sequence of dates
dates = pd.date_range('01-01-2023', periods=41, freq='D')

#creating pandas Series with date index
s = pd.Series(dates)
print (s)
```
